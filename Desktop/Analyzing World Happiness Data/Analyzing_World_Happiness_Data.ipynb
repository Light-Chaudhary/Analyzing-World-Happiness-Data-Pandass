```python
import pandas as pd

url = "https://raw.githubusercontent.com/turdubars/World-Happiness/master/2015.csv"
df = pd.read_csv(url)
print(df.head())

#Dropping Some Columns

df = df[["Country", "Region", "Happiness Rank", "Happiness Score", "Family" ,"Economy (GDP per Capita)", "Health (Life Expectancy)"]]

# Rename columns to simpler names: e.g., Country, HappinessScore, GDP, Support, LifeExpectancy, etc.

df.rename(columns = {"Happiness Score": "HappinessScore", "Family": "Support", "Economy (GDP per Capita)": "GDP", "Health (Life Expectancy)" : "LifeExpectancy"}, inplace = True)

# # Sort and show Top 10 happiest countries:

top10_happiest_countries = df.nlargest(10, "HappinessScore")
print(top10_happiest_countries)

# Filter countries with high GDP but low happiness.

average_gdp = df["GDP"].mean()
average_happiness = df["HappinessScore"].mean()

filtered_countries = df[(df['GDP'] > average_gdp) & (df["HappinessScore"] < average_happiness)]
print(filtered_countries)

# Compute correlation between Happiness Score and other factors (GDP, Support, etc.).

print(df[['HappinessScore', 'GDP', 'Support']].corr())

# Identify bottom 10 least happy countries.

least10_happy_countries = df.sort_values("HappinessScore", ascending = True).head(10)
print(least10_happy_countries)

# Export your top 10 and bottom 10 tables to CSV:
top10_happiest_countries.to_csv('top10_happiest_country.csv', index = False)
least10_happy_countries.to_csv('least10_happy_countries.csv', index = False)
```

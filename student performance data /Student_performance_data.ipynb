```python
import pandas as pd

url = "https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv"
df = pd.read_csv(url)
df.head()
print(df.columns.tolist())

# Rename columns to lowercase and more readable names (e.g., "Height(Inches)" → height, "Weight(Pounds)" → weight)
df.rename(columns = {' Height(Inches)"' : 'Height',  ' "Weight(Pounds)"' : 'Weight'}, inplace = True)

# Check for any missing values and handle them appropriately
print(df[df.isnull().any(axis = 1)])

# Sort by height in descending order
df.sort_values("Height", ascending = False, inplace = True)

# Reset the index after sorting
df.reset_index(drop = True, inplace = True)

# Add a new column rank based on height (1 = tallest)
df["Rank"] = df["Height"].rank(ascending = False , method = "dense")

# Move the rank column to be second (after height)
df = df[["Height", "Rank", "Weight"]]

#  Filter and show top 10 tallest people
top10_tallest_people = df.nlargest(10, "Height")
print(top10_tallest_people)

# Show average height and weight
average_height = df["Height"].mean()
print(average_height)
average_weight = df["Weight"].mean()
print(average_weight)

#  Filter people with weight above average
filtered_people = df[df["Weight"] > average_weight]
print(filtered_people)

# Display only those whose height is below average but weight is above average
print(df[(df["Height"] < average_height) & (df["Weight"] > average_weight)])
```
